```python
x = dataset.iloc[:,:-1]
y = dataset["Selling_Price"]
```

```python
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import root_mean_squared_error, mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test, lr.predict(x_test)), root_mean_squared_error(y_test, lr.predict(x_test))
```

```python
lr1 = Lasso(alpha=.1)
lr1.fit(x_train,y_train)
lr1.score(x_train,y_train)*100, lr1.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test, lr1.predict(x_test)), root_mean_squared_error(y_test, lr1.predict(x_test))
```

```python
lr2 = Ridge(alpha=.1)
lr2.fit(x_train,y_train)
lr2.score(x_train,y_train)*100, lr2.score(x_test,y_test)*100
```

```python
mean_squared_error(y_test, lr2.predict(x_test)), root_mean_squared_error(y_test, lr2.predict(x_test))
```

```python
df = { "criterion" : ["squared_error", "friedman_mse", "absolute_error"],
       "splitter" : ["best", "random"],
       "max_depth" : [i for i in range(2,30)]
     }
         
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train, y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

```python
dt = DecisionTreeRegressor(criterion= 'squared_error', max_depth= 28, splitter= 'best',min_samples_split=2,min_samples_leaf=1)
dt.fit(x_train,y_train)
dt.score(x_train, y_train)*100, dt.score(x_test, y_test)*100
```

```python
mean_squared_error(y_test, dt.predict(x_test)), root_mean_squared_error(y_test, dt.predict(x_test))
```

```python
rv = RandomForestRegressor(n_estimators=150,criterion='squared_error',min_samples_split=2,max_depth=8)
rv.fit(x_train,y_train)
rv.score(x_train, y_train)*100, rv.score(x_test, y_test)*100
```

```python
mean_squared_error(y_test, rv.predict(x_test)), root_mean_squared_error(y_test, rv.predict(x_test))
```
