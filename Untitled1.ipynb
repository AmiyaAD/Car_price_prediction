```python
def train_test_split(data, test_ratio):
    np.random.seed(42)
    suffeled = np.random.permutation(len(data))
    
    print(suffeled)
    test_set_size = int(len(data) * test_ratio)
    train_indexes = suffeled[test_set_size:]
    test_indexes = suffeled[:test_set_size]
    return data.iloc[test_indexes], data.iloc[train_indexes]
```

```python
x_t, y_t = train_test_split(data, 0.2)
```
